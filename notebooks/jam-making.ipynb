# Eventual Boiling Points for Sucrose Solutions

In this notebook we investigate boiling points of sucrose solutions as a function of altitude and solution concentration. It is known from [Browne 1912](http://archive.org/details/handbookofsugara00browrich) that the boiling point of a sucrose solution may be represented as a fixed elevation of temperature from the boiling point of pure water. The boiling point of pure water depends on the ambient pressure and, hence, elevation.

## Boiling point of water

### Analysis

[Snyder](http://www.hi-tm.com/Documents/Calib-boil.html) gives the following formulae for computing pressure, $P$, from elevation, $h$, and boiling point, $T$, from pressure:

$$ P_{hg} = 29.921 \times (1-6.8753 \times 10^{-6} \times h_{ft})^{5.2559} \quad \mbox{and} \quad T_{F} = 49.161 \times \ln(P_{hg}) + 44.932 $$

Unfortunately these are in units, feet, inches of mercury and degrees Fahrenheit, which are of little interest to the UK based cook. Fortunately there are some simple conversions into the more useful units of metres, millibars and degrees centigrade.

$$ P = 33.864 \  P_{hg} \quad h = 0.3048 \  h_{ft} \quad T = \frac{T_{F} - 32}{1.8} $$

Folding those into the equations above gives:

$$ P = 1013.24 \times (1-2.2556 \times 10^{-5} \times h)^{5.2559} \quad \mbox{and} \quad T = 27.312 \times \ln(P) - 89.02 $$

Let us translate these equations into Python while also importing the `pylab` computation system:

```python
%pylab inline

def P(h):
    return 1013.24 * (1 - 2.2566e-5 * h) ** 5.2559

def T(P):
    return 27.312 * np.log(P) - 89.02
```

### Validation

We can test these functions by plotting some graphs of pressure versus altitude and boilind point versus pressure.

```python
altitudes = np.linspace(-500, 2000, 256)
plot(altitudes, P(altitudes))
title('Atmospheric pressure versus altitude')
axvline(0, color='r', label='Sea level')
axhline(1013.25, color='g', label='Mean pressure at sea-level')
legend()
xlabel('Altitude above sea-level [m]')
ylabel('Pressure [millibar]')
grid('on')
```

```python
pressures = np.linspace(940, 1060, 256)
plot(pressures, T(pressures))
axvline(1013.25, color='r', label='Mean pressure at sea-level')
axhline(100, color='g', label=u'100\u00B0C')
title('Boiling point of water versus atmospheric pressure')
legend(loc='upper left')
xlabel('Pressure [millibar]')
ylabel(u'Temperature [\u00B0C]')
grid('on')
```

Notice that both graphs are consistent with a) the definition of 100 degrees centigrade being the boiling point of water at sea level and b) the mean atmospheric pressure at se level being known to be 1013.25 millibar. This provides some validation that our formulae are correct.

## Boiling point of a sucrose solution

We start with the table as presented in [Browne 1912] with alternate columns showing sucrose percentage and the boiling point electation in degrees centigrade.

```python
browne_t = array([[  2. ,   0.3,  32. ,   1.1,  62. ,   3.4],
                  [  4. ,   0.3,  34. ,   1.2,  64. ,   4. ],
                  [  6. ,   0.3,  36. ,   1.3,  66. ,   4.8],
                  [  8. ,   0.4,  38. ,   1.4,  68. ,   5.6],
                  [ 10. ,   0.4,  40. ,   1.5,  70. ,   6.5],
                  [ 12. ,   0.4,  42. ,   1.6,  72. ,   7.4],
                  [ 14. ,   0.5,  44. ,   1.7,  74. ,   8.4],
                  [ 16. ,   0.5,  46. ,   1.8,  76. ,   9.4],
                  [ 18. ,   0.5,  48. ,   1.9,  78. ,  10.6],
                  [ 20. ,   0.6,  50. ,   2. ,  80. ,  12. ],
                  [ 22. ,   0.7,  52. ,   2.1,  82. ,  13.9],
                  [ 24. ,   0.7,  54. ,   2.3,  84. ,  16.3],
                  [ 26. ,   0.8,  56. ,   2.4,  86. ,  19.3],
                  [ 28. ,   0.9,  58. ,   2.7,  88. ,  23.1],
                  [ 30. ,   1. ,  60. ,   3. ,  90. ,  27.8]])
```

This is of little use, however, since we'd prefer the table stacked into a single two column matrix:

```python
bpe_t = np.vstack((browne_t[:,0:2], browne_t[:,2:4], browne_t[:,4:6]))
```

Using the `scipy` interpolation routines, we can convert this into a boiling point elevation function `bpe`:

```python
from scipy import interpolate
bpe = interpolate.interp1d(bpe_t[:,0], bpe_t[:,1])
```

We can now plot the boiling point elevation against sucrose solution:

```python
solutions = np.linspace(2, 70, 256)
plot(solutions, bpe(solutions))
title('Boiling point elevation of versus sucrose solution')
xlabel('Solution strength [%]')
ylabel(u'Boiling point elevation [\u00B0C]')
grid('on')
```

## Combined tables

We now want to construct an image representing the boiling point against both solution and pressure.

```python
pressures = np.linspace(1013-30, 1013+30, 1024)
solutions = np.linspace(60-15, 60+15, 1024)

# what size should our image be?
n_rows = pressures.shape[0]
n_cols = solutions.shape[0]
extent=(solutions.min(), solutions.max(), pressures.max(), pressures.min())

# create arrays of pressure and solution indices
p_idxs, s_idxs = np.indices((n_rows, n_cols))

# create arrays of pressures and solution strengths
ps = pressures[p_idxs]
ss = solutions[s_idxs]

# create array of water boiling points
water_bps = T(ps)

# create array of BPEs
bpes = bpe(ss)

# create combined array
bps = water_bps + bpes
```

```python
figure(figsize=(11,8))
clabel(contour(solutions, pressures, bps, levels=np.arange(100,150,0.5), colors='k'), fmt='%.1f')
title('Boiling temperature for sugar solutions')
xlabel('Sugar solution strength [%]')
ylabel('Atmospheric pressure [millibar]')
grid('on')
savefig('jam-making-figure.pdf')
```

### Verification

When making jam one aims for a 60% solution and usually one uses a boiling temperature of 103 degrees centigrade (http://www.exploratorium.edu/cooking/icooks/01-06-03.html). Looking at our figure we see this corresponds to the expected mean sea-level pressure of around 1013 millibar.

On the BBC weather site (http://bbc.co.uk/weather) the pressure at the time and location of writing is unusually low at around 1000 millibar. On days such as today the author would boild his solution to around 102.5 in order to get the required 60% solution.
